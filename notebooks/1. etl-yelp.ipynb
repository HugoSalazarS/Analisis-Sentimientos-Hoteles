# Yelp dataset cleaning

```python
#Cargamos las librer√≠as necesarias para hacer la limpieza
import pandas as pd
import os
```

```python
# Asignamos el path relativo a la variable dir para utilizarla a la hora de consumir los datasets
# Obtener el path actual
dir = os.getcwd()
```

```python
# Cargamos el dataset 'business.parquet' que contiene la lista de negocios
df_yelp_business = pd.read_parquet(dir+'/Yelp/parquets/business.parquet')
```

```python
df_yelp_business.head()
```

```python
df_yelp_business.info()
```

## Creamos tablas con filtros en la columna 'categories' y 'name'

### Filtro en la columna 'name'

```python
# Eliminamos filas con valores faltantes (NaN)
df_yelp_business = df_yelp_business.dropna(subset=['name'])

#Utilizamos str.lower() para convertir el texto a min√∫sculas y luego str.contains()
filtro_inn = df_yelp_business['name'].str.lower().str.contains('inn|hotel|suite')

# Aplicamos el filtro para obtener las filas que contienen 'hotel' o 'inn' o 'suite' en la columna 'name'
filas_con_inn = df_yelp_business[filtro_inn]
```

```python
filas_con_inn.head()
```

```python
filas_con_inn.shape
```

### Filtro en las columnas 'name' y 'categories'

```python
# Eliminamos filas con valores faltantes (NaN)
filas_con_inn = filas_con_inn.dropna(subset=['categories'])

#Utilizamos str.lower() para convertir el texto a min√∫sculas y luego str.contains()
filtro_hotel_inn = filas_con_inn['categories'].str.lower().str.contains('hotel')

# Aplicamos el filtro para obtener las filas que contienen 'hotel' en la columna 'categories'
filas_con_hotel_inn = filas_con_inn[filtro_hotel_inn]
```

```python
filas_con_hotel_inn.head()
```

```python
filas_con_hotel_inn.shape
```

## Creamos una tabla con filtro en la columna 'categories' unicamente

```python
# Eliminamos filas con valores faltantes (NaN)
df_yelp_business = df_yelp_business.dropna(subset=['categories'])

#Utilizamos str.lower() para convertir el texto a min√∫sculas y luego str.contains()
filtro_hotel = df_yelp_business['categories'].str.lower().str.contains('hotel')

# Aplicamos el filtro para obtener las filas que contienen 'hotel'
filas_con_hotel = df_yelp_business[filtro_hotel]

# Restablecemos el √≠ndice y eliminar la columna "index"
filas_con_hotel.reset_index(inplace=True, drop=True)
```

```python
filas_con_hotel.head()
```

```python
filas_con_hotel.shape
```

## Trabajamos sobre el dataset checkin

```python
df_yelp_checkin = pd.read_parquet(dir+'/Yelp/parquets/checkin.parquet')
```

```python
df_yelp_checkin.head()
```

```python
# Creamos un dataframe para agregar la columna'date' renombrandola a 'checin_date'
checkin_merge = df_yelp_checkin.rename(columns={'date': 'checkin_date'})
```

```python
# Utilizamos el m√©todo merge() para agregar la columna 'checkin_date' de df_yelp_checkin a filas_con_hotel basado en la columna 'business_id'
df_final = filas_con_hotel.merge(checkin_merge[['business_id', 'checkin_date']], on='business_id', how='left')


df_final.head()
```

# Trabajamos sobre el dataset review

```python
df_yelp_review = pd.read_parquet(dir+'/Yelp/parquets/review.parquet')
```

```python
df_yelp_review.head()
```

```python
df_yelp_review.info()
```

```python
# Creamos un dataframe para agregar la columna'date' renombrandola a 'checkin_date'
review_merge = df_yelp_review.rename(columns={'date': 'review_date', 'text': 'review_text'})
```

```python
review_merge.head()

```

```python
# Creamos una tabla 'review_merge' con las filas filtradas seg√∫n los valores de la columna 'business_id' de df_final
# Obtener los valores √∫nicos de la columna "business_id" del DataFrame df_final
valores_filtro = df_final['business_id'].unique()

# Filtrar las filas de review_merge basado en los valores de "business_id" que coincidan con "business_id" en df_final
review_merge = review_merge[review_merge['business_id'].isin(valores_filtro)]

# Restablecemos el √≠ndice y eliminar la columna "index"
review_merge.reset_index(inplace=True, drop=True)
```

```python
review_merge.info()
```
